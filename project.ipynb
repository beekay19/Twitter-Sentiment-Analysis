```python
import twint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nest_asyncio
import re
from textblob import TextBlob
from nltk.tokenize import WordPunctTokenizer
from nltk.corpus import stopwords   
from datetime import datetime
import glob                     
import os
nest_asyncio.apply()
```

```python
name=['Tinubu','Atiku','Peter Obi']
```

```python
def twintConfig(search_string):
    c = twint.Config()
    c.Search = search_string
    c.Limit=5000
    c.Pandas = True
    c.Pandas_au = True          
    c.Pandas_clean=True
    twint.run.Search(c)
```

```python
def Run_Twint(search_vals):
    
    #set empty dataframe for join
    out_df= pd.DataFrame()
    
    for politician in search_vals:
        print ("running for search item: "+politician +"\n")
        #run twint
        twintConfig(politician)
        
        #get dataframe
        tweets_df = twint.storage.panda.Tweets_df
        
        #join Dataframes and create 'Bank' column
        tweets_df["Politician"]= politician
        out_df = pd.concat([out_df,tweets_df])
        
    return out_df
```

```python
tweets_df  = Run_Twint(name)
```

```python
def clean_text(text):  
    pat1 = r'@[^ ]+'                   #@signs
    pat2 = r'https?://[A-Za-z0-9./]+'  #links
    pat3 = r'\'s'                      #floating s's
    pat4 = r'\#\w+'                     # hashtags
    pat5 = r'&amp '
    pat6 = r'[^A-Za-z\s]'         #remove non-alphabet
    combined_pat = r'|'.join((pat1, pat2,pat3,pat4,pat5, pat6))
    text = re.sub(combined_pat,"",text).lower()
    return text.strip()
```

```python
tweets_df["tweet"]   =  tweets_df['tweet'].apply(clean_text)
```

```python
print("Running sentiment process")
# creating two new columns(polarity and subjectivity)
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity

tweets_df["Subjectivity"] = tweets_df['tweet'].apply(getSubjectivity)
tweets_df["Polarity"] = tweets_df['tweet'].apply(getPolarity)
```
